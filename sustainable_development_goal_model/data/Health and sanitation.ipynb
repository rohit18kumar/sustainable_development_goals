```python
import pandas as pd
import numpy as np
from scipy.stats import norm,pearsonr
import io
from scipy import stats
```

```python
df=pd.read_csv('Health and sanitation.csv')
df=df.loc[0:35]
df
```

```python
mean = 74
```

```python
df = df.sort_values('Clean water and sanitation',ascending = False)
new_df =df.head(5)
new_df
```

```python
import scipy
GndEq=df['Health']
t_value,p_value = scipy.stats.ttest_1samp(GndEq,mean,alternative='greater')
print('Test statistic is %f'%float("{:.6f}".format(t_value)))

print('p-value for one tailed test is %f'%p_value)
```

```python
import scipy
GndEq=df['Health']
t_value,p_value = scipy.stats.ttest_1samp(GndEq,mean,alternative='less')
print('Test statistic is %f'%float("{:.6f}".format(t_value)))

print('p-value for one tailed test is %f'%p_value)
```
